```python
import pandas as pd
```

```python
# load dataset
df = pd.read_csv("./data/movie_metadata.csv")
print(df.shape)
df.head()
```

```python
# drop rows with NaN
df = df.dropna()
print(df.shape)
```

```python
# drop duplicated rows
df = df.drop_duplicates()
print(df.shape)
```

```python
# check data types of each column
df.dtypes
```

```python
# select columns of desired data types
data = df.select_dtypes(exclude=['object'])
data.dtypes
```
